```python
from simul_gbm import simul_gbm
from Calc_Payoff import payoff
from dataClass_Phoenix import PhoenixProduct
import numpy as np

%run dataClass_Phoenix.ipynb
%run simul_gbm.ipynb
%run Calc_Payoff.py

def phoenix_price(product, S0, r, sigma, nb_simulations):
    paths = simul_gbm(S0, r, sigma, product.observation_dates, nb_simulations)
    pv_list = []
    for path in paths :
        pv, t = payoff(path, product, S0)
        pv_actu = pv* np.exp(-r*t)
        pv_list.append(pv_actu)
    price = np.mean(pv_list)
    return price

    
```

```python
product = PhoenixProduct(maturity=5.0, coupon_barrier=0.70, protection_barrier=0.60, coupon_rate=0.08)
price = phoenix_price(product, S0=100, r=0.03, sigma=0.20, nb_simulations=10000)
print(f"Prix Phoenix : {price:.2f} €")
```
